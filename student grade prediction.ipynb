Courses like HS-101 and CS-105 seem to have students performing relatively well, while subjects like PH-121, MT-111, and CS-312 may need more attention in terms of student support or curriculum adjustment to reduce lower grades and withdrawals.

```python
#Data Prerprocessing
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
for i in df.columns:
    if df[i].dtypes == 'object':
        df[i] = oe.fit_transform(df[i].values.reshape(-1, 1))
df.head()

```

```python
df.hist(bins=10,figsize=(15,15));
```

Some courses like HS-101, PH-121, and CS-105 show data that is more evenly distributed across the X-axis, suggesting a wider range of performance among students.
Other courses, such as EE-119, MT-111, and CS-212, show more skewed distributions, where a large number of students fall into specific bins (either higher or lower), potentially reflecting that most students performed similarly in these subjects.

The final chart shows the CGPA distribution for all students, with most students having a CGPA between 2.5 and 3.5. This suggests that the majority of students are performing within an average to above-average range.
